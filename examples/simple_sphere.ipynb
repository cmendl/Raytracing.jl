# Render a simple sphere

```julia
using Images
using FileIO
```

```julia
include("../src/Raytracing.jl")
using .Raytracing
```

```julia
# image dimensions (pixel)
nx = 200
ny = 100;
```

```julia
# define camera
lookfrom = zeros(3)
lookat = [0., 0., -1.]
vup = [0., 1., 0.]
vfov = π/2
aperture = 0
focus_dist = 1
camera = Camera(lookfrom, lookat, vup, vfov, nx / ny, aperture, focus_dist);
```

```julia
# define scene geometry

mat = Lambertian([0.5, 0.5, 0.5])

scene = SurfaceAssembly([])
push!(scene.objects, Sphere([0.,    0.,  -1.], 0.5, mat))
# large "globe" sphere imitating ground floor
push!(scene.objects, Sphere([0., -100.5, -1.], 100, mat));
```

```julia
# number of samples (rays) per pixel
ns = 100

img = renderimage(nx, ny, ns, scene, camera)

# show rendered image
transpose(img)
```

```julia
# save to disk
save("simple_sphere.png", transpose(img))
```
